```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = sns.load_dataset('tips')
# print(data)
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "day", size="size", marker = "o", palette = "spring")
plt.show()
```

```python
data = pd.read_excel("ESD.xlsx")
# print(data)
sns.scatterplot(data = data, x = "Age", y = "Annual Salary", size = "Bonus %", hue = "Department")
"""plt.legend(bbox_to_anchor=(1.4,1.2))  # it can also contain 4 parameters"""
plt.legend(bbox_to_anchor=(0.2,0,1.2,1))  # (x, y, width, height)
plt.savefig('scatterplot.png')
plt.show()
```
