```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset('tips')
# print(data)
gp = data.groupby('day').agg({"tip":"mean"})
# print(gp)
sns.heatmap(gp)
plt.show()
```

```python
data = pd.read_excel("ESD.xlsx")
gp = data.groupby("Job Title").agg({"Annual Salary":"mean"})
sns.heatmap(gp)
plt.show()
```
