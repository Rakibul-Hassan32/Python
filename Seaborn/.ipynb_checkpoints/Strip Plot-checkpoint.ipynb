```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = sns.load_dataset("tips")
sns.stripplot(data=data, x = "day", y = "total_bill", hue = "sex", dodge = True, jitter = 0.3)  # data shifts = jitter, seperate two type data = dodge
plt.legend(bbox_to_anchor = (1.3,1))
plt.show()
# good for all points, if the data is not numerical doesn't matter
```
