```python
import pandas as pd
df = pd.DataFrame({'Name': ['Jane','Michel','William','Rosy','Hana','Fredie','Graeme'],
                  'Salary': [50000,54000,50000,189000,55000,40000,59000]})
print(df)
```

```python
# calculate the mean
print(df['Salary'].mean())  # gor
```

```python
# median, mode
print(df['Salary'].median())  # moddhok
print(df['Salary'].mode())  # prochurok
print(df['Salary'].min())
print(df['Salary'].max())
```

```python
df['Salary'].describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
print(df)
salary= df['Salary']
salary.plot.hist(title='Salary Distribution', color='gray',bins=25)
plt.axvline(salary.mean(),color='violet',linestyle='dashed',linewidth=2)
plt.axvline(salary.median(),color='red',linestyle='dashed',linewidth=2)
plt.show()
```
