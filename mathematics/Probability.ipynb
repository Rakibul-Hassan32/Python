```python
from itertools import product as prod
A = set([x for x in range(1,7)])
B = set([x for x in range(1,7)])
p = list(prod(A,B))
print(p)
```

```python
# throwing 2 dice
n_dice = 2
dice_faces = {1,2,3,4,5,6}
event_space = set(prod(dice_faces,repeat=n_dice))
event_space
```

```python
for outcome in event_space:
    print(outcome,end=', ')
print()
print(len(event_space))
```

```python
# sum is a multiple of 3
favourable_outcome = []
for outcome in event_space:
    x,y = outcome
    if(x+y)%3==0:
        favourable_outcome.append(outcome)
favourable_outcome = set(favourable_outcome)
print(favourable_outcome)    
print(len(favourable_outcome))

prob = len(favourable_outcome)/len(event_space)
print(prob)
```

```python
# throwing 5 dice
n_dice = 5
dice_faces = {1,2,3,4,5,6}
event_space = set(prod(dice_faces,repeat=n_dice))
event_space
```

```python
len(event_space)
```

```python
# sum is a multiple of 5 but not 3
favourable_outcome = []
for outcome in event_space:
    d1,d2,d3,d4,d5 = outcome
    if (d1+d2+d3+d4+d5) % 5 == 0 and (d1+d2+d3+d4+d5) % 3 != 0:
        favourable_outcome.append(outcome)

favourable_outcome = set(favourable_outcome)
print(favourable_outcome)
print(len(favourable_outcome))

prob = len(favourable_outcome) / len(event_space)
print(prob)
```
