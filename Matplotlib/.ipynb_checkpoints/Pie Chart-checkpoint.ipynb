```python
import matplotlib.pyplot as plt
brands = ["OnePlus","Apple","Samsung","Nokia","Redmi"]
x = [22, 35, 30, 3, 10]
c = ["red","silver","blue","pink","orange"]
ex = [0,0.1,0,0,0.1]
plt.pie(x, labels = brands,colors = c, explode = ex, shadow = "True", autopct = "%.2f",startangle = 90)  # explode is used to highlight a pie
plt.savefig("pie.png",facecolor = "green",pad_inches=2,bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)
groupedby = df.groupby("Payment Mode")["Amount"].sum()
print(groupedby)
ex = [0.2,0, 0]
plt.pie(groupedby.values, labels = groupedby.index, autopct = "%.2f", shadow = True, explode = ex)
plt.show()
# print(df)
```
