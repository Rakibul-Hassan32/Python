```python
import matplotlib.pyplot as plt
y = [98,67,88,95,88]
x = ["Part1","Part2","Part3","Part4","Part5"]
colors = ["red","green"]
plt.bar(x , y, color=colors,edgecolor="black")
plt.xlabel("Parts of Harry Potter", fontsize=17)
plt.ylabel("Popularity", fontsize=17)
plt.title("Popularity of Harry Potter Series", fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)
# print(df)
groupedby = df.groupby("Payment Mode")["Amount"].sum()
print(groupedby)
plt.bar(groupedby.index,groupedby.values)
plt.savefig("bar.png")  # to save in the folder
plt.show()
```
