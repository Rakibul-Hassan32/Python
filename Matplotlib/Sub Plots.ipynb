```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [45, 34, 56, 23, 45]

plt.subplot(2,2,1)  # row, column, chart number
plt.bar(x,y)

x = [5,6,7,8,9]
y = [67,50,66,56,82]

plt.subplot(2,2,2)
plt.scatter(x,y)

x = [2,4,6,8,10]
y = [57,50,60,55,20]

plt.subplot(2,2,3)
plt.step(x,y,marker="o",where="mid")

x = [1,3,7,5,9]
y = [67,51,62,50,23]

plt.subplot(2,2,4)
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [45, 34, 56, 23, 45]

plt.subplot(1,2,1)  # row, column, chart number
plt.title("Age")
plt.bar(x,y)

x = [5,6,7,8,9]
y = [67,50,66,56,82]

plt.subplot(1,2,2)
plt.title("Weight")
plt.plot(x,y)
plt.suptitle("Employee Data")  # super title
plt.show()
```
