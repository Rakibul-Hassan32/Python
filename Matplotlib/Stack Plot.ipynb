```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]
NOP1 = [5,10,30,20,35,60,80]
NOP2 = [50,60,30,75,80,90,120]
NOP3 = [8,20,50,60,70,90,100]

plt.stackplot(days,NOP1,NOP2,NOP3,colors = ["pink","hotpink","magenta"],labels=["week1","week2","week3"])
plt.legend(loc = 2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel("samplefood.xlsx")
df = pd.DataFrame(data)
# print(df)
days = [1,2,3,4,5,6,7]
a = ["Calories","Protien","Fat"]
grouped = df.groupby("Category")[a].agg("mean")
print(grouped)
plt.stackplot(df["Category"].unique(), grouped["Calories"],grouped["Protien"],grouped["Fat"],labels=("Calories","Protein","Fat"))
plt.legend(loc = 2)
plt.show()
```
